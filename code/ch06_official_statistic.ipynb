### e-Stat APIをPythonで利用する


````python
from urllib.parse import urljoin
import requests
import pandas as pd
````

````python
import os
from dotenv import load_dotenv
load_dotenv()
appId = os.getenv("ESTAT_APP_ID")
version = "3.0"
base_url = f"https://api.e-stat.go.jp/rest/{version}/"

print(base_url)
````

````python
# APIキーの動作確認
import requests # まだインポートしていなければ

# 1. そもそも .env から appId が読み込めているか？
if appId:
    print(f"1. .env読み込み: 成功 (先頭: {appId[:5]}...)")
else:
    print("1. .env読み込み: 失敗 (変数が空です。 .env ファイルの変数名が ESTAT_APP_ID になっているか確認してください)")

# 2. 実際に e-Stat API を叩いて認証を確認する
test_url = f"{base_url}app/json/getStatsList"  # さきほどの base_url を活用
test_params = {
    "appId": appId,
    "statsCode": "00200521" # 家計調査のテスト用コード
}

try:
    res = requests.get(test_url, params=test_params)
    data = res.json()
    
    # e-Statからのレスポンス内にあるステータスを確認
    # JSONの階層が深いので get() で安全に取得します
    status = data.get("GET_STATS_LIST", {}).get("RESULT", {}).get("STATUS")

    if status == 0:
        print("2. 認証確認: 成功！(正しく通信できています)")
    else:
        error_msg = data.get("GET_STATS_LIST", {}).get("RESULT", {}).get("ERROR_MSG")
        print(f"2. 認証確認: 失敗 (エラー: {error_msg})")
        
except Exception as e:
    print(f"2. 認証確認: 通信エラーが発生しました ({e})")
````

#### 「接頭辞（頭の文字）」のルール
これらはAPIの仕様書で「こういう条件で絞り込みたい時は、こう名乗ってね」と決まっています。

- `cd` (Codeの略): 「〜のコード番号で絞り込む」という意味（例: `cdCat02`）
- `lv` (Levelの略): 「階層の深さで絞り込む」という意味（例: `lvCat01`）
- `Cat01`, `Cat02`...: これは、さきほど内包表記で確認した 「分類項目の番号」 です。

#### 例：lvCat01": "4"
1. どこから？: 前のステップの `enumerate` の結果で、0番目が「表章項目(tab)」、1番目が「用途分類(cat01)」 だと分かりましたね。
2. 意味: 「1番目の分類（用途分類：Cat01）について、階層（lv）が 4（最も細かい内訳） のデータだけをください」という注文です。

#### 例：cdCat02": "04"
1. どこから？: `metadata["CLASS_INF"]["CLASS_OBJ"][2]["CLASS"]`（2番目の分類＝世帯区分のリスト）を表示させると、その中に コード "04"、名前 "勤労者世帯" という項目が見つかります。
2. 意味: 「2番目の分類（世帯区分：Cat02）は、コード（cd）が 04番 のデータだけをください」という注文です。

````python
value_df = pd.json_normalize(data, 
    record_path=["GET_STATS_DATA", "STATISTICAL_DATA", "DATA_INF", "VALUE"]
)
value_df.head()
````

````python
value_df = value_df.rename(
    columns=lambda col: col.lstrip("@").replace("$", "value")   # lstripで @ を除去し、replaceで $ を value に置き換えている
)
````

````python
note = data["GET_STATS_DATA"]["STATISTICAL_DATA"]["DATA_INF"]["NOTE"]
note
````

出力に `{'@char': '***', '$': '調査又は集計していないもの'}` などとあるが、 pythonの辞書はカンマで区切れば1つの `{ }` の中にいくつでもペアを入れることができる

````python
import numpy as np

# 1. 【リスト内包表記】
# noteから注釈記号（@char）だけを抜き出してリスト化する
# 普通のforループを1行にまとめた「リストの量産」命令
# note_char = ["-", "***", ...] のようになる
note_char = [n["@char"] for n in note]

# 2. 【assign ＋ lambda式】
# データフレームに列を追加・更新する。
# lambda df: ... は「現在のデータフレームを df として受け取って加工する」という、その場限りの使い捨て関数
value_df = value_df.assign(**{
    "value": lambda df: df["value"]
        # 数字以外の記号（注釈）を、計算可能な「欠損値(NaN)」に置き換える
        .replace(note_char, np.nan)
        # 列全体を、計算ができるように「文字列(str)」から「実数(float)」に変換する
        .astype(float)
})


# forループの普通の書き方（やってることは全く同じ）
# note_char = []           # 1. まず空のリストを用意する
# for n in note:           # 2. noteの中身を1つずつ取り出すループを回す
#     item = n["@char"]    # 3. その中から "@char" という値だけ抜き出す
#     note_char.append(item) # 4. 用意したリストに付け加える

````

assign と lambda の組み合わせは、Pandasの中でも特に「慣れるまでが大変な、でも分かると便利な」書き方です。

それぞれの役割を、さらに細かく分解して解説します。

---

1. assign の書き方の意味

`value_df = value_df.assign( 新しい列名 = 加工内容 )` というのが基本形です。

##### 謎の記号 `**{ }` は何？

このコードでは `**{"value": ...}`という少し特殊な書き方をしています。

- 本来なら `assign(value = lambda ...)` と書けばいいのですが、たまに列名に「@」などの記号が含まれていると、Pythonの文法エラー（変数名として使えない）になってしまいます。
- そこで、「辞書型で作った設定を、無理やり割り当てる」 ための呪文として **{ } が使われています。今回の場合、`value` は普通の名前なので、単に `assign(value = ...)` と書いても同じ意味です。

##### `assign` の最大の特徴

「元のデータ（value_df）を直接いじらず、新しく作り替えたデータのコピーを作成する」という性質があります。だから、一番最初に value_df = と書いて、新しい自分自身で上書きしているわけです。

---
2. lambda の書き方の意味
なぜ単純に `df["value"]` と書かずに、`lambda df: df["value"]` と書く必要があるのでしょうか？

##### `df` はどこから来たのか？
この lambda の右側に書いてある df は、実は 「今、assign されている最中の value_df そのもの」 が自動的に代入されてきます。

##### もし `lambda` を使わなかったら？

```python
# これはエラーになるか、古いデータを参照してしまう
value_df.assign(value = value_df["value"].replace(...))
```

これだと、「assign が始まる前」の古い value_df を見にいってしまいます。

##### `lambda` を使うと何が良い？

```python
# これなら最新の自分を参照できる
value_df.assign(value = lambda df: df["value"].replace(...))
```

lambda を使うことで、Pandasが「はい、今のデータはこれだよ！」と df という器に中身を移して渡してくれます。 これにより、もしこの前に別の加工（例えば列の削除など）が済んでいたとしても、その「最新の状態」に対して加工を続けられるのです。

---

3. 全体の流れを会話にすると……
1. 左側 (value_df = value_df.assign) 「value_df 君、君のコピーを新しく作るよ。結果はまた自分自身に保存してね」
2. 中の辞書 ("value": ...) 「新しく作り直す列の名前は value にしよう」
3. ラムダ (lambda df: ...) 「今の君の姿を一時的に df と呼ぶね。その中の value 列を取り出して……」
4. 加工 (.replace().astype()) 「記号を消して、数値（小数）に変換してちょうだい」


---
##### なぜこのセットで書くのか（結論）

この書き方をすると、1つも「途中の変数（ゴミ）」を作らずに、元のデータを安全に最新の状態へアップデートできるからです。

最初は lambda df: df の部分を、「今まさに加工しているデータを捕まえておくための網（あみ）」 だと考えておけば、イメージが掴みやすいかもしれません！_

````python
from typing import List, Dict, Union
def missing_to_nan(
        value: pd.DataFrame, 
        note: Union[Dict[str, str], List[Dict[str, str]]]
    ) -> pd.DataFrame:
    if isinstance(note, list):
        note_char = [n["@char"] for n in note]
    elif isinstance(note, dict):
        note_char = note["@char"]
    else:
        print(f"引数noteの型は辞書またはリスト。noteの型: {type(note)}")
        return value
    return value.assign(**{
        "value": lambda df: df["value"]
            .replace(note_char, np.nan)
            .astype(float)
    })
````
